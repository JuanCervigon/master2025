```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=400, noise=0.05, random_state=0)
```

```python
x = X[:, 0]
y = X[:, 1]
```

```python
import seaborn as sns
import pandas as pd

# Create a Pandas DataFrame from your x and y data
data = pd.DataFrame({'x': x, 'y': y})

# Use the 'x' and 'y' columns as keyword arguments in scatterplot
sns.scatterplot(x='x', y='y', data=data, legend=False);
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 0.2)
```

```python
clusters = model.fit_predict(X)
```

```python
import seaborn as sns

# Assuming 'data' is your DataFrame and 'clusters' is your cluster assignments
data['clusters'] = clusters  # Add the cluster assignments to your DataFrame

# Use the 'x' and 'y' columns from the DataFrame and 'clusters' for hue
sns.scatterplot(x='x', y='y', hue='clusters', data=data);
```
